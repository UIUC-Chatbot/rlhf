## Hand-written QA from google docs -> CSV formatting

```python
def parse_file(file_path):
  questions = []
  answers = []
  with open(file_path, 'r') as f:
    lines = f.readlines()
    i = 0
    while i < len(lines):
      if lines[i].startswith('Question:'):
        current_question = lines[i].replace('Question:', '').strip()
        current_answer = lines[i + 1].replace('Answer:', '').strip()
        questions.append(current_question)
        answers.append(current_answer)
        i += 2
      else:
        i += 1
  return questions, answers


questions, answers = parse_file('raw_student_written_qa.txt')
```

# Combine these new data with exsiting RLHF data

```python
import datasets

existing_dataset = datasets.load_dataset("kastan/rlhf-qa-conditional-generation")
new_dataset = datasets.load_dataset('csv', data_files='cleaned_student_written_QA.csv')

print(existing_dataset)
print(new_dataset)
```

```python
dataset = datasets.concatenate_datasets([existing_dataset['train'], new_dataset['train']])
print("Combined datasets (before dedup)\n",dataset)

# remove duplicates
import pandas as pd
df = pd.DataFrame(dataset)
df.drop_duplicates(subset=['prompt', 'completion'], inplace=True)
dataset = datasets.Dataset.from_pandas(df)
print("Combined datasets (AFTER dedup)\n",dataset)
```
