```python
import re

def clean_chapter_title(my_string):
  '''
  my_string = "1.2 The 2's Complement Representation . . . . . . . . . . . . . . . . . . . . . . . . . . 3"
  out = "The 2's Complement Representation"
  '''

  # Step 1: remove the leading digits and dots
  # pattern1 = r'^\d+\.\d+\s'
  pattern1 = r'^\d+(\.\d+)* '
  cleaned_string1 = re.sub(pattern1, '', my_string)
  # print(cleaned_string1)

  # Step 2: remove the trailing dots and spaces
  # pattern2 = r'\s*\.\s*[\d\s]*$'
  pattern2 = r' \.(\s*\.){2,}\s*\. \d*'
  cleaned_string2 = re.sub(pattern2, '', cleaned_string1)
  print(cleaned_string2.rstrip('\n'))
  return cleaned_string2.rstrip('\n')
```

```python
# open full_coverage_lumetta_raw.txt and read it line by line
# and write it to full_coverage_lumetta_raw.csv
with open('full_coverage_lumetta_raw.txt', 'r') as f:
  lines = f.readlines()
  # f2.write(line.replace('\t', ','))


# <A>
# dd
# defd
# </A?

from collections import defaultdict
import re

pattern = r'\d+\.\d+\s+(\S.*\S)\s*\.\s*\d'
# r'\d+\.\d+\s+(\S.*\S)\s*\.\s*'


# {chapter, question, answer}
questions_only = []
data = []
curr_data = defaultdict(str)
# curr_answer = ''

for i in range(len(lines)):
  if i <=9:
    # skip header
    continue

  if lines[i] == '<CHAPTER>\n':
    curr_data['chapter'] = clean_chapter_title(lines[i+1])
    i += 2
  elif lines[i] == '<QUESTION>\n':
    
    if curr_data['prompt']:
      # question doesn't have an answer, so we save it to questions_only.
      curr_chapter = curr_data['chapter']
      curr_data['completion'] = '' 
      questions_only.append(curr_data)
      curr_data = defaultdict(str)
      curr_data['chapter'] = curr_chapter
      
    
    curr_data['prompt'] = lines[i+1].rstrip('\n')
    
    i += 1
  elif lines[i] == '<ANSWER>\n':
    i += 1
    while lines[i] != '</ANSWER>\n':
      curr_data['completion'] += lines[i]
      i += 1
    i += 2
    curr_data['completion'] += curr_data['completion'].rstrip('\n')
    data.append(curr_data)
    curr_data = defaultdict(str)

  # if i >40:
  #   break

# print(data)

# for i in data:
#   print(i)
```
